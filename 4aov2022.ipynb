```python
input_path = r"C:\Users\work\aov2022\aoc\input4.txt"

with open(input_path,'r') as a:
    uncleaneddata = a.readlines()
```

```python
data = [x[:-1] for x in uncleaneddata]
```

```python
count = 0
for ids in data:
    values = [int(x) for x in ids.replace('-',',').split(',')]
    if values[2] >= values[0] and values [3] <= values[1]:
        count+=1
    elif values[2] <= values[0] and values [3] >= values[1]:
        count+=1
    else:
        pass




```
