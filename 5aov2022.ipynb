```python
def move(column_from=7, column_to=2, amount=2, columns_values=columns):
    column_from_index = column_from-1
    column_to_index = column_to-1

    print(f'to column: {column_to_index}, {columns_values[column_to_index]}'
          f'from column : {column_from_index}, {columns_values[column_from_index]}')

    crates_to_send = []
    for x, crate in enumerate(columns_values[column_from_index]):

        while amount > 0 and len(columns_values[column_from_index]) > 0:
            print(f'amount: {amount}, '
                  f'crate popped: {columns_values[column_from_index][x]}')
            crates_to_send.append(columns_values[column_from_index].pop(x))
            amount -= 1


    for moved_crate in crates_to_send:
        columns_values[column_to_index].insert(0, moved_crate)

    print(f'new column: {columns_values[column_to_index]}')
    print(f'old column: {columns_values[column_from_index]}')

    return columns_values


```
