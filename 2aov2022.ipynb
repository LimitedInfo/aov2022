```python
with open(input_path,'r') as a:
    guide = a.readlines()
```

```python
guide = [x[:-1] for x in guide]
```

```python
game_map = {'A': 'Rock', 'B': 'Paper', 'C': 'Scissors', 'X': 'Rock', 'Y': 'Paper', 'Z': 'Scissors'}
```

```python
def game_result(opp,you):
    # print(opp,you)
    if opp==you:
        return 0.5
    if opp=='Rock' and you == 'Paper':
        return 1
    elif opp=='Rock' and you =='Scissors':
        return 0
    elif opp=='Paper' and you =='Rock':
        return 0
    elif opp=='Paper' and you == 'Scissors':
        return 1
    elif opp=='Scissors' and you =='Rock':
        return 1
    elif opp=='Scissors' and you == 'Paper':
        return 0
```

```python
total_score = 0
for game in guide:
    # print(total_score)
    total_score += game_score(game_result(game_map[game[0]],game_map[game[-1]]),game_map[game[-1]])
```

```python
def game_score(result,option):
    print(result, option)
    if result == 1:
        game_result_score = 6
    elif result == 0.5:
        game_result_score = 3
    elif result == 0:
        game_result_score = 0

    if option == 'Rock':
        return 1 + game_result_score
    elif option == 'Paper':
        return 2 + game_result_score
    elif option == 'Scissors':
        return 3 + game_result_score
```

```python
def what_to_throw(opp,desired_game_result):
    print('opp {}'.format(opp))
    print(f'desired_game_result {desired_game_result}')
    if opp=='Rock' and desired_game_result=='Win':
        return 'Paper'
    elif opp=='Rock' and desired_game_result=='Draw':
        return 'Rock'
    elif opp=='Rock' and desired_game_result=='Loss':
        return 'Scissors'
    elif opp=='Paper' and desired_game_result=='Win':
        return 'Scissors'
    elif opp=='Paper' and desired_game_result=='Draw':
        return 'Paper'
    elif opp=='Paper' and desired_game_result=='Loss':
        return 'Rock'
    elif opp=='Scissors' and desired_game_result=='Win':
        return 'Rock'
    elif opp=='Scissors' and desired_game_result=='Draw':
        return 'Scissors'
    elif opp=='Scissors' and desired_game_result=='Loss':
        return 'Paper'
```
